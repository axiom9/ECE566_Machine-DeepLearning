##Part 2: Central Limit Theorem

Prove the Central Limit Theorem through a simple experiment.

Compare experiment with the theoretical equivalent gaussian curve.

More information: https://en.wikipedia.org/wiki/Central_limit_theorem

```python
### Suggestion: consider the rolls of a die.

### Experiment: 
# Roll a die 1000 times and record the sample mean of those 1000 rolls, repeat this 5, 10, 50, 100, 10000 times to effectively see the CLT in action

# die roll can be simulated using np.random.choice, considering the normal 6 sided die.

def simulate(n):

    sample_means = []

    for _ in range(n):
        sample = np.random.choice(6, 1000)
        sample_means.append(np.mean(sample))

    
    plt.hist(sample_means)
    plt.title(f'Sample means for {n} trials')
    plt.xlabel('Sample mean')
    plt.ylabel('Count')

### Theoretical gaussian curve:

#Theoretical mean? variance?
### Representation
```

```python
simulate(n=5)
```

```python
simulate(n=10)
```

```python
simulate(n=50)
```

```python
simulate(n=100)
```

```python
simulate(n=10000)
```

Evidently, we can see that as the 'n' (the amount of times we repeat the experiment) increases, the sample mean becomes closer and closer to a normal distribution (gaussian). That's the entire basis of the central limit theorem, the sample means will portray a normal distribution.
